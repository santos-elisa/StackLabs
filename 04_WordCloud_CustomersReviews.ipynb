# **Customers' Reviews Word Cloud Analysis**

**Description**

[Olist](https://olist.com/pt-br/) is a Brazilian Unicorn that offers e-commerce solutions for small and mid-size companies in Brazil.

The Olist datasets are related to Sales between 2017 and 2018 in many categories, from Bed Bath & Table to Agro.

These datasets can act as a source for various analysis, from Sales Prediction to Customer Satisfaction.

**Objective**

Creating a word cloud to support the analysis of 14447 customers' reviews as below:

* all product categories;
* products delivered in 2017 and 2018;
* review scores between 1 and 3, of a maximum of 5;
* reviews written in Brazilian Portuguese.

**Source**

Olist Datasets: https://www.kaggle.com/olistbr/brazilian-ecommerce

```python
# Importing libraries.
import os
from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
# Creating a data directory using getcwd() to support the module generate in Python.
d = path.dirname(__file__) if "__file__" in locals() else os.getcwd()
```

```python
# Importing the txt file with the customers' reviews.
text = open(path.join(d, '/content/drive/MyDrive/StackLabs2201/Datasets/olist_customers_reviews.txt')).read()
```

# **The word cloud**

```python
# Generating the word cloud image.
wordcloud = WordCloud(max_font_size=40, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

# **Analysis**

 The Order Reviews dataset contains 14447 entries with scores between 1 and 3, of a maximum of 5.

 It is possible to observe that most of the words are related to delivery.

 Stopwords were not removed from the dataset because they may be relevant in this context.

 Words such as "no" or "nor" combined with terms like "delivered" or "received" can represent that the customer did not receive the product or received it after the deadline.
